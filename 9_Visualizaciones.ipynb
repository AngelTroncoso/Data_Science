#Visualizaciones con Python: Matplotlib, Searbon y Plotly

```python
import pandas as pd

df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df.head()
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc['Colombia', anos]
colombia.head()
```

Creamos un diccionario para guardar esta variable

```python
col_dict = {'Año':colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist()}
col_dict
```

Convertir diccionario en Dataframe

```python
datos_col = pd.DataFrame(col_dict)
datos_col.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980', '1990', '2000', '2010'])
plt.show()
```

```python
argentina_brasil = df.loc[['Argentina', 'Brasil'], anos]
argentina_brasil = argentina_brasil.T
argentina_brasil.head()
```

```python
argentina_brasil.index = argentina_brasil.index.map(int)
argentina_brasil.plot(kind='line', figsize=(10, 6))
plt.title('Inmigración de Argentina y Brasil a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.xticks(range(1980, 2014, 5)) # Mostrar ticks cada 5 años
plt.show()
```
